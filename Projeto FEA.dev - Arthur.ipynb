```python
#Importando bibliotecas Pandas,Matplotplib e Seaborn
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Definindo as variáveis dos arquivos CSV

url = 'https://stgdwsngpcpd01.blob.core.windows.net/dados-abertos/EDA_Manipulados_201707.csv'
url2 = 'https://stgdwsngpcpd01.blob.core.windows.net/dados-abertos/EDA_Manipulados_201807.csv'
url3 = 'https://stgdwsngpcpd01.blob.core.windows.net/dados-abertos/EDA_Manipulados_201907.csv'
url4 = 'https://stgdwsngpcpd01.blob.core.windows.net/dados-abertos/EDA_Manipulados_202007.csv'
url5 = 'https://stgdwsngpcpd01.blob.core.windows.net/dados-abertos/EDA_Manipulados_202107.csv'

df = pd.read_csv(url, encoding = 'cp1252', sep = ';')
df2 = pd.read_csv(url2, encoding = 'cp1252', sep = ';')
df3 = pd.read_csv(url3, encoding = 'cp1252', sep = ';')
df4 = pd.read_csv(url4, encoding = 'cp1252', sep = ';')
df5 = pd.read_csv(url5, encoding = 'cp1252', sep = ';')
```

```python
#Utilizando a variável tabela_unica para concatenar os 2 CSV em uma única planilha

tabela_unica = pd.concat([df,df2,df3,df4,df5], axis=0, join='inner')
```

```python
#based on https://stackoverflow.com/questions/17071871/how-do-i-select-rows-from-a-dataframe-based-on-column-values
drugs = ['ZOLPIDEM',
        'CLONAZEPAM',
        'BROMAZEPAM',
        'LORAZEPAM',
        'FLUOXETINA',
        'ESCITALOPRAM',
        'MIRTAZAPINA',
        'DULOXETINA',
        'IVERMECTINA',
        'CLOROQUINA',
        'HIDROXICLOROQUINA',
        'AZITROMICINA']

df_drugs = tabela_unica[tabela_unica['PRINCIPIO_ATIVO'].isin(drugs)]
```

```python
#Definindo os gráficos dos remédios Antidepressivos

fig = plt.figure(figsize=(15, 15))
tips = sns.load_dataset('tips')
sns.set_style('whitegrid')

ax1 = fig.add_subplot(221)
sns.countplot(data=df_drugs.loc[df_drugs.PRINCIPIO_ATIVO == drugs[4]], x='ANO_VENDA').set_title(drugs[4])
ax1.set_xlabel('Anos', fontsize=13)
ax1.set_ylabel('Quantidade', fontsize=13)

ax2 = fig.add_subplot(222)
sns.countplot(data=df_drugs.loc[df_drugs.PRINCIPIO_ATIVO == drugs[5]], x='ANO_VENDA').set_title(drugs[5])
ax2.set_xlabel('Anos', fontsize=13)
ax2.set_ylabel('Quantidade', fontsize=13)

ax3 = fig.add_subplot(223)
sns.countplot(data=df_drugs.loc[df_drugs.PRINCIPIO_ATIVO == drugs[6]], x='ANO_VENDA').set_title(drugs[6])
ax3.set_xlabel('Anos', fontsize=13)
ax3.set_ylabel('Quantidade', fontsize=13)

ax4 = fig.add_subplot(224)
sns.countplot(data=df_drugs.loc[df_drugs.PRINCIPIO_ATIVO == drugs[7]], x='ANO_VENDA').set_title(drugs[7])
ax4.set_xlabel('Anos', fontsize=13)
ax4.set_ylabel('Quantidade', fontsize=13)
```
